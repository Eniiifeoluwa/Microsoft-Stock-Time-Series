```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Microsoft_Stock.csv")
dataset.set_index('Date', inplace= True)
```

```python
dataset.plot(figsize= (30, 10))
plt.xticks(rotation = 90);
```

```python
def plot_hist(data):

    fig = plt.figure(figsize= (20, 8))
    for i, col in enumerate(data.columns):
        ax1 = fig.add_subplot(1, len(data.columns), i+1, xticks = [], yticks = [])
        color = ['red', 'blue', 'green', 'black', 'purple']
        ax1.hist(data[col], color = color[i] )
        ax2 = fig.add_subplot(2, len(data.columns), i+1, xticks = [], yticks = [])
        ax2.plot(data[col], color = color[i])

        plt.title(col)
    plt.show();
plot_hist(dataset)
```

```python

print("Statistical Summary:")
display(dataset.describe())


print("\nMissing Values:")
display(dataset.isnull().sum())


plt.figure(figsize=(8, 6))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python

sns.pairplot(dataset)
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()


fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
for i, col in enumerate(dataset.columns):
    sns.histplot(dataset[col], ax=axes[i], kde=True, color='skyblue')
    axes[i].set_title(f'Distribution of {col}')
for j in range(len(dataset.columns), 6):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
dataset.plot(kind='box', subplots=True, layout=(1,5), figsize=(18,6), patch_artist=True)
plt.suptitle("Boxplots for Outlier Detection")
plt.show();

close_df = dataset['Close']
plt.figure(figsize=(20, 6))
close_df.plot(label='Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Microsoft Stock Close Price Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show();
```
